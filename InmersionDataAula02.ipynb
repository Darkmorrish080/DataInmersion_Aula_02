```python
import pandas as pd
#import desde drive del csv a usar
inmuebles = pd.read_csv('/content/drive/MyDrive/DataInmersion/inmuebles_bogota.csv')
inmuebles.sample(8)
```

```python
inmuebles.info()
```

```python
inmuebles.Valor[0]
```

```python
#tratamiento para poder sumar la columna valor
inmuebles.Valor[0].split()
```

```python
valor = inmuebles.Valor.str.split(expand=True)
inmuebles['Moneda'] = valor[0]
inmuebles['Precio'] = valor[1]
inmuebles.sample(3)
```

```python
inmuebles.info()
```

```python
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','',regex=True)
```

```python
inmuebles[['Precio','Barrio']]
```

```python
inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000
inmuebles.info()
```

```python
inmuebles.describe()
```

```python
pd.set_option('display.precision',2)
pd.describe_option('display.float_format',lambda x: '%.2f' % x)
inmuebles.describe()
```

```python
inmuebles.loc[inmuebles.Habitaciones == 110] #Localizar un dato 
```

```python
inmuebles['Precio_Millon'].plot.hist(bins=100)
```

```python
#seaborn para visualizar datos estadisticos mas detallados

import matplotlib.pyplot as plt #otra lib para estadistica
import seaborn as sns

plt.figure(figsize=(10,6)) #definir tamanio del lienzo
grafica = sns.histplot(data=inmuebles, x='Precio_Millon',kde=True, hue='Tipo')
grafica.set_title('Distribucion de valores de los inmuebles en Bogota')
plt.xlim((50,1000))
plt.savefig('/content/drive/MyDrive/DataInmersion/valor_inmuebles_bogota.png',format='png')
plt.show()
```
